```python
## Antes de começar a gerar os relatórios por aqui é bom validar que o input principal, os códigos de produto
## estarão certos para todos os produtos do período.

from bling import Api, ApiError
import pandas as pd

api = Api('920feadbcc9525902531f5ec7c67c53ddb8cd79e7a4dc03a00cf48087f8389c866f877ca')

itens_pedidos = []
##pedidos em andamento - BOLETOS JÁ PAGO E CARTÃO.
try:
	pedidos = api.get_orders(issued_date=['01/01/2021', '31/01/2021']) ##periodo dos pedidos selecionados
	for pedido in pedidos:
		if pedido['situacao'] == 'Em andamento':
			for itens in (pedido['itens']):
				itens_pedidos.append(itens)
except ApiError as e:
	print(e.response)
```

```python
#extrai só os codigos dos itens incluidos nos pediddos
codigos_contagem=[]
desc_contagem=[]
qtd=[]

for x in range(len(itens_pedidos)):
    codigos_contagem.append(itens_pedidos[x]['item']['codigo'])
    desc_contagem.append(itens_pedidos[x]['item']['descricao'])
    qtd.append(itens_pedidos[x]['item']['quantidade'])
```

```python
#converte a lista em DF
import pandas as pd

DF_pedidos = pd.DataFrame(list(zip(codigos_contagem,desc_contagem,qtd)),columns = ['codigo','desc','qtd'])
```

```python
DF_pedidos.to_excel('camisetas.xlsx')
```

```python
#separa em tres colunas com o divisor -
divisao = DF_pedidos['codigo'].str.split('-')
codigo = divisao.str.get(0)
tamanho = divisao.str.get(1)
cor = divisao.str.get(2)
DF_pedidos['codigo'] = codigo
DF_pedidos['tamanho'] = tamanho
DF_pedidos['cor'] = cor
```

```python
#faz depara de P e B pra preto e branco
for index, row in  DF_pedidos.iterrows():
    if row['cor'] == 'P':
        DF_pedidos.loc[index,'cor'] = 'Preto'
    elif row['cor'] == 'B':
        DF_pedidos.loc[index,'cor'] = 'Branco'
```

```python
#Removendo colunas que não tem preto ou branco na cor - na teoria ficam só as camisetas (se cadastro tiver correto)
DF_camisetas = DF_pedidos[(DF_pedidos['cor'] == 'Preto') | (DF_pedidos['cor'] == 'Branco')].reset_index()
DF_camisetas.drop(columns = 'index', axis = 1, inplace = True)
```

```python
DF_camisetas.head(40)
```

```python
#Criando DF das regatas femininas tradicionais e salvando a planilha (pro Antonio) e a separação (pro Ricardo)
df_retr = DF_camisetas.loc[DF_camisetas['codigo'].str[0:4] == 'RETR'].reset_index(drop=True)
df_retr['qtd'] = df_retr['qtd'].astype(float)
df_retr.groupby(['desc']).agg({'qtd':'sum'}).rename(columns={"qtd":"quantidade"}).sort_values('desc',ascending=False).reset_index()
df_retr.to_excel('1.2regatasfemininastradicionaissilk.xlsx')
df_retr['tamcor'] = df_retr['tamanho']+ " " + df_retr['cor']
separacao_antonio_retr = df_retr.groupby(['tamcor']).agg({'qtd':'sum'}).reset_index()
separacao_antonio_retr = separacao_antonio_retr.rename(columns={"tamcor":"tamanho/cor"})
separacao_antonio_retr = separacao_antonio_retr.sort_values('qtd',ascending=False).reset_index(drop=True)
separacao_antonio_retr.to_excel('2.2separacao_antonio_regatas_femininas_tradicionais.xlsx',
                               sheet_name = 'REGATASFEMININASTRADICIONAIS')
```

```python
#Criando DF das regatas masculinas e salvando a planilha (pro Antonio) e a separação (pro Ricardo)
df_rema = DF_camisetas.loc[DF_camisetas['codigo'].str[0:4] == 'REMA'].reset_index(drop=True)
df_rema['qtd'] = df_rema['qtd'].astype(float)
df_rema.groupby(['desc']).agg({'qtd':'sum'}).rename(columns={"qtd":"quantidade"}).sort_values('desc',ascending=False).reset_index()
df_rema.to_excel('1.3regatasmasculinassilk.xlsx')
df_rema['tamcor'] = df_rema['tamanho']+ " " + df_rema['cor']
separacao_antonio_rema = df_rema.groupby(['tamcor']).agg({'qtd':'sum'}).reset_index()
separacao_antonio_rema = separacao_antonio_rema.rename(columns={"tamcor":"tamanho/cor"})
separacao_antonio_rema = separacao_antonio_rema.sort_values('qtd',ascending=False).reset_index(drop=True)
separacao_antonio_rema.to_excel('2.3separacao_antonio_regatas_masculinas.xlsx',
                               sheet_name = 'REGATASMASCULINAS')
```

```python
#Criando DF das regatas femininas long e salvando a planilha (pro Antonio)
df_relo = DF_camisetas.loc[DF_camisetas['codigo'].str[0:4] == 'RELO'].reset_index(drop=True)
df_relo['qtd'] = df_relo['qtd'].astype(float)
df_relo.groupby(['desc']).agg({'qtd':'sum'}).rename(columns={"qtd":"quantidade"}).sort_values('desc',ascending=False).reset_index()
df_relo.to_excel('1.4regataslongsilk.xlsx')
df_relo['tamcor'] = df_relo['tamanho']+ " " + df_relo['cor']
separacao_antonio_relo = df_relo.groupby(['tamcor']).agg({'qtd':'sum'}).reset_index()
separacao_antonio_relo = separacao_antonio_relo.rename(columns={"tamcor":"tamanho/cor"})
separacao_antonio_relo = separacao_antonio_relo.sort_values('qtd',ascending=False).reset_index(drop=True)
separacao_antonio_relo.to_excel('2.4separacao_antonio_regatas_femininas_long.xlsx',
                               sheet_name = 'REGATASFEMLONG')
```

```python
#Excluindo as regatas do DF original das camisetas
regatas = ['REMA', 'RETR', 'RELO']
df_camisetas = DF_camisetas[~DF_camisetas.codigo.str[0:4].isin(regatas)]
df_camisetas['qtd'] = df_camisetas['qtd'].astype(float)
```

```python
df_camisetas.reset_index(drop=True)
df_camisetas.to_excel('camisetassilkcodigo.xlsx')
```

```python
#agrupa por modelo, conta e salva (pro Antonio)
df_camisetas_silk = df_camisetas.groupby(['codigo']).agg({'qtd':'sum'}).rename(columns={"codigo":"quantidade"}).sort_values('codigo',ascending=False).reset_index()
df_camisetas_silk.to_excel('1.1camisetassilk.xlsx')
```

```python
#Itens que não são camisas
#agrupa por modelo, conta, converte em float e salva (pro Antonio)
DF_outros = DF_pedidos[(DF_pedidos['cor'] != 'Preto') & (DF_pedidos['cor'] != 'Branco')].reset_index(drop=True)
DF_outros.drop(columns = ['tamanho','cor'], inplace = True)
DF_outros['qtd'] = DF_outros['qtd'].astype(float)
DF_outros_silk = DF_outros.groupby(['desc']).agg({'qtd':'sum'}).rename(columns={"qtd":"quantidade"}).sort_values('desc',ascending=False).reset_index()
DF_outros_silk.to_excel('1.5outrossilk.xlsx')
```
